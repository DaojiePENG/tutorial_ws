## 2.1 Python Introduction

Python 是一种高级编程语言，结合了解释性、编译性、互动性和面向对象的特性。它以简洁易读的语法设计而闻名，非常适合初学者和专业开发者。

Python 的主要特点包括：

* 解释型语言：无需编译，代码可以直接运行，类似于 PHP 和 Perl。

* 交互式：支持在终端输入代码并立即执行，方便测试和调试。

* 面向对象：支持面向对象编程，允许代码封装在对象中。

* 易学易用：语法简单，关键字少，代码结构清晰。

* 跨平台：支持在 UNIX、Windows 和 Mac 等多个平台运行。

* 丰富的标准库：内置大量模块，支持文件操作、网络编程、数据库访问等功能。

* 可扩展性：可以通过 C 或 C++ 编写高性能模块，并在 Python 中调用。

* 支持 GUI：可以开发图形用户界面应用程序。

* 嵌入性：可以嵌入到其他语言（如 C/C++）的程序中，提供脚本化能力。

Python 由 Guido van Rossum 于 20 世纪 80 年代末设计，受多种语言（如 ABC、C、SmallTalk 等）启发。它的开源特性和强大的社区支持使其成为全球最流行的编程语言之一，广泛应用于数据科学、人工智能、Web 开发、自动化等领域。
